```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
df=pd.read_csv("/content/Advertising.csv")
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
sns.pairplot(df,x_vars=['TV','Radio','Newspaper'],y_vars='Sales',kind='scatter')
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',linewidth=0.5)
plt.show()
```

```python
plt.hist(df['TV'],bins=20,color='skyblue', edgecolor='black')
plt.xlabel('TV')
```

```python
plt.hist(df['Radio'],bins=20,color='Green', edgecolor='black')
plt.xlabel('Radio')
```

```python
df['Newspaper'].plot.hist(bins=30,color='Purple',edgecolor='black',xlabel='Newspaper')
plt.xlabel('Newspaper')
```

```python
fig = px.sunburst(df, path=['TV','Radio','Newspaper'], values='Sales',color='Sales')
fig.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['TV','Radio','Newspaper']],df[['Sales']], test_size=0.2, random_state=0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
res=model.predict(X_test)
print(res)
```

```python
plt.plot(res)
```

```python
mae = mean_absolute_error(y_test, res)
print("Mean Absolute Error:", mae)
```

```python
new_data = pd.DataFrame({
  'TV': [100],
    'Radio': [20],
    'Newspaper': [30]
})
predicted_sales = model.predict(new_data)
print("Predicted Sales:", predicted_sales)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,res)
```
